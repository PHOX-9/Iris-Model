```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Species']=le.fit_transform(df['Species'])
```

```python
setosa=df[df['Species']==0]
versicolor=df[df['Species']==1]
verginica=df[df['Species']==2]
```

```python
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(setosa['SepalLengthCm'],setosa['SepalWidthCm'],color='green',marker='+')
plt.scatter(versicolor['SepalLengthCm'],versicolor['SepalWidthCm'],color='blue',marker='.')
plt.scatter(verginica['SepalLengthCm'],verginica['SepalWidthCm'],color='red',marker='*')
```

```python
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(setosa['PetalLengthCm'],setosa['PetalWidthCm'],color='green',marker='+')
plt.scatter(versicolor['PetalLengthCm'],versicolor['PetalWidthCm'],color='blue',marker='.')
plt.scatter(verginica['PetalLengthCm'],verginica['PetalWidthCm'],color='red',marker='*')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df.head()
```

```python
X=df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
Y=df['Species']
```

```python
X
```

```python
Y
```

```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
```

```python
scaler=StandardScaler()
Xscale=scaler.fit_transform(Xtrain)
Xscaletest=scaler.transform(Xtest)
```

```python
knn.fit(Xscale,Ytrain)
```

```python
knn.score(Xscaletest,Ytest)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
Ypredict=knn.predict(Xscaletest)
```

```python
#confusion_matrix(truth,prediction)
cm=confusion_matrix(Ytest,Ypredict)
```

```python
cm
```

```python
import seaborn as sns
plt.figure(figsize=(7,5))
sns.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
import joblib
joblib.dump(knn,'Iris Species.pkl')
joblib.dump(scaler,'Scaler.pkl')
```
